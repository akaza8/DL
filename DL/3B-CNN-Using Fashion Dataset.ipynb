```python
from tensorflow.keras.datasets import fashion_mnist
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
(train_x, train_y), (test_x, test_y) = fashion_mnist.load_data()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, MaxPooling2D, Conv2D
```

```python
model = Sequential()
```

```python
model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu',input_shape=(28, 28, 1)))

# Adding maxpooling layer to get max value within a matrix
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(128, activation = "relu"))
model.add(Dense(10, activation = "softmax"))

model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(train_x.astype(np.float32), train_y.astype(np.float32), epochs = 15, validation_split = 0.2)
```

```python
predictions = model.predict(test_x)
```

```python
class_names=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal','Shirt', 'Sneaker', 'Bag', 'Ankleboot']
```

```python
import numpy as np
index=10
print(predictions[index])
final_value=np.argmax(predictions[index])
print("Actual label :",test_y[index])
print("Predicted label :",final_value)
print("Class :",class_names[final_value])
```

Evaluating the model

```python
loss, accuracy = model.evaluate(test_x, test_y)
print("Loss :",loss)
print("Accuracy (Test Data) :",accuracy*100)
```
